## A. Evaluation Procedure Number 1: Train and Test on the entire dataset.

### A.1. USING SUPPORT VECTOR MACHINE

###### A.1.a To Create the Support Vector Machine Model

```python
#To Import the Support Vector Machine Model
from sklearn.svm import SVC

#To Instantiate the Model (Using Majority of Default Parameters)
support_vector_machine = SVC(kernel = 'linear', random_state=0)

#To Fit the Whole Dataset into the Support Vector Machine Model
support_vector_machine.fit(X,Y) #All Dataset

#To Predict the Output of the Whole Dataset
Y_predict_SVM = support_vector_machine.predict(X)
Y_predict_SVM

#To Show the Histogram of the Predicted Classification
plot.hist(Y_predict_SVM, bins=None)
plot.xlim(0,2)
plot.title('Histogram of Predicted Species Classification')
plot.xlabel('Species Classification')
plot.ylabel('Frequency')
```

###### A.1.b To Evaluate the Performance of the Support Vector Machine Model

```python
# To Show the Confusion Matrix, to compare true value and predicted value of Y
from sklearn.metrics import confusion_matrix
print('CONFUSION MATRIX:')
print(confusion_matrix(Y,Y_predict_SVM)) #Y - true value ; Y_predicted_SVM - predicted results of the model
```

```python
# For the Classification Accuracy by comparing true value and predicted value of Y
from sklearn.metrics import accuracy_score
classification_accuracy = accuracy_score(Y,Y_predict_SVM)
print('Classification Accuracy: %.4f'
      % classification_accuracy)
print('')
```

```python
# For the Classification Report to get the precision, recall, F1-score, and support
from sklearn.metrics import classification_report
print ('CLASSIFICATION REPORT: ')
print(classification_report(Y,Y_predict_SVM))
```
