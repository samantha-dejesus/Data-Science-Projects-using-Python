```python
#Calculate the MEAN, and replace any empty values with it:
data4 = pd.read_csv('Data.csv')
x=data4["Salary"].mean()
xx=data4["Age"].mean()
data4["Salary"].fillna(x, inplace=True)
data4["Age"].fillna(xx, inplace=True)
print(data4.to_string())
```

```python
data5 = pd.read_csv('Data.csv')
```

```python
#Calculate the MEDIAN, and replace any empty values with it:
C=data5["Age"].median()
D=data5["Salary"].median()
data5["Age"].fillna(C,inplace=True)
data5["Salary"].fillna(D,inplace=True)
print(data5.to_string())
```

```python
#Calculate the MODE, and replace any empty values with it:
data6 = pd.read_csv('Data.csv')
E=data6["Age"].mode()[0]
F=data6["Salary"].mode()[0]
data6["Age"].fillna(E, inplace = True)
data6["Salary"].fillna(F, inplace = True)
print(data6.to_string())
```

# SCALE FEATURES / FEATURE SCALING

```python
#Scale all values in the Weight and Volume Columns:
#Import Libraries
import pandas
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scale
```

```python
#Independent Variable
X1=data_co2[['Weight','Volume']]
X1
```

```python
#Dependent Variable
y=data_co2['CO2']
y
```

```python
#Feature Scale the Independent Variable
scaledX1=scale.fit_transform(X1)
print(scaledX1)
```

```python
#Predict the CO2 emission from a 1.3 liter car that weighs 2300 kilograms:
mlr = linear_model.LinearRegression()
mlr.fit(scaledX1,y)
```

```python
scaled=scale.transform([[2300,1.3]])
predictedCO2=mlr.predict([scaled[0]])
print(predictedCO2)
```
